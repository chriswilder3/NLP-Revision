**1. Countplot**


The `seaborn.countplot()` function is used to show the counts of observations in each categorical bin using bars. It's a helpful tool for visualizing the distribution of a categorical variable.

Here are some important parameters:

*   **`x`, `y`**: These parameters determine which variable in your dataset you want to count. You provide the name of the column from your DataFrame. You can use either `x` or `y` to specify the categorical variable. For example, `x='Sex'` will create a countplot with 'Sex' on the x-axis.
*   **`data`**: This is the DataFrame containing the data you want to plot. You pass your DataFrame object here (e.g., `data=df`).
*   **`hue`**: This parameter allows you to add another categorical variable to the plot, creating separate bars for each category within the main variable. This is useful for comparing the distribution of the main variable across different groups. For example, `hue='Survived'` would show the count of each 'Sex' category, broken down by whether they survived or not.
*   **`order`**: This parameter allows you to specify the order in which the bars are displayed on the axis. You can pass a list of the categories in the desired order.
*   **`palette`**: This parameter lets you customize the colors of the bars. You can use predefined color palettes or provide a list of colors.

**Objects that can be passed to `countplot`:**

*   **Pandas Series:** As you did in your example `sns.countplot(df['Survived'].value_counts())`, you can pass a Pandas Series containing the counts of each category. However, it's more common and often easier to pass the DataFrame and specify the column using `x` or `y`.
*   **Pandas DataFrame:** This is the most common way to use `countplot`. You pass the entire DataFrame to the `data` parameter and specify the column(s) using `x`, `y`, and `hue`.
*   **Numpy arrays:** You can also pass NumPy arrays, but using DataFrames is generally preferred when working with structured data.

**Different ways it can display data:**

*   **Vertical bars:** By default, `countplot` will create vertical bars if you specify the categorical variable using `x`.
*   **Horizontal bars:** If you specify the categorical variable using `y`, `countplot` will create horizontal bars.
*   **Grouped bars:** When you use the `hue` parameter, the bars will be grouped together based on the categories in the `hue` variable.

Here's an example using the `data`, `x`, and `hue` parameters:

```python
# Example using data, x, and hue

sns.countplot(x='Sex', hue='Survived', data=df)

  # Provide col_name as either x or y. Passing both is not allowed. hue is inner category.
```

**Examples :**

```python
# Survived Col

# We need to pass the dataframe and colname as x

sns.countplot(x = 'Survived', data = df)
```

```python
# We can do the similar with Pandas bar plot as well
print(df["Survived"].value_counts())
df["Survived"].value_counts().plot(kind = "bar")
```

```python
# Pclass

sns.countplot(x= "Pclass", data = df, palette = "muted") # Palette is for colors ex : muted, deep, pastel, hls
```

# PieChart

It also does frequency counts for each class, but it displays percentages instead of count.

IMP : Seaborn itself does not have a dedicated pie chart function, we can use matplotlib’s pie() function

```python
import matplotlib.pyplot as plt
```

 A. Simple Pie Chart

```python
plt.pie(x=df["Embarked"].value_counts())
```

```python
# We can see that its too simple, lets add labels and other details, but first we need to obtain labels using Pandas

df["Embarked"].value_counts()
```

```python
# To retrieve the index (or "row names") of a Pandas DataFrame, you can directly access the .index attribute of the DataFrame object.

df["Embarked"].value_counts().index # we still need to convert it to list
```

```python
labels = list(df["Embarked"].value_counts().index)
```

 B. PieChart with labels

```python
labels = list(df["Embarked"].value_counts().index)

plt.pie(x = df["Embarked"].value_counts(), labels = labels)
```

 C. Pie Chart with labels and percentages

```python
# change autopct to strings to '%1.1f%%'

plt.pie(x = df["Embarked"].value_counts(), labels = labels, autopct = "%1.2f%%")
```

 D. Pie Chart explode, shadow, and rotate slices

```python
# Make explode, array/list/tuple indicating which segment should be split apart like a cake(exploded) and by how much
explode_vals = (0,0.2,0) # always better < 1

plt.pie(x = df["Embarked"].value_counts(), explode = explode_vals)
```

```python
# Add shadow with shadow = True

plt.pie(x = df["Embarked"].value_counts(), shadow = True)
```

```python
# customize starting angle with startangle= 90
plt.pie(x = df["Embarked"].value_counts(), startangle = 180)
```

```python
# We can Combine SNS palettes with MATPLOTLIB's plots, 
# NOT ALL palettes are supported

plt.pie(x = df["Embarked"].value_counts(), colors=sns.color_palette("deep") )
```

```python
# we can plot simple pie bar using pandas plot( kind = "pie")

df["Embarked"].value_counts().plot(kind = "pie")
```

**2. Histogram**

We can split the data range into bins, display count of each bin, making groups for continous data. It is hence useful in converting numerical data to categorical.

- It shows us the distribution of data like Skewness and shape.

We can use 3 ways to histogram

1. Matplotlib's plt.hist(x , bins = 3, range= (10,90))

2. panda's .plot(kind = "hist") (internally uses matplotlib)

3. seaborn's sns.histplot(x,y, data etc)

**NOTE :** If not specified, Matplotlib choses 10 bins & seaborn 20 bins


1. A) Matplotlib's `plt.hist()` & pandas `df.plot(kind="hist")`

This is the foundational function for creating histograms in Matplotlib. It's flexible and gives you fine-grained control over the histogram's appearance.

**Key Arguments:**

*   **`x`**: The data you want to plot. This should be a numerical array or list.
*   **`bins`**: This determines how the data range is divided. It can be:
    *   An integer: The number of equal-width bins.
    *   A sequence (list or array): The edges of the bins. This gives you more control over the bin ranges.
*   **`range`**: The lower and upper bounds of the bins. If not specified, the range is determined by the minimum and maximum values in the data.
*   **`density`**: If `True`, the histogram will be normalized to form a probability density, meaning the area under the histogram integrates to 1. This is useful for comparing distributions with different sample sizes.
*   **`weights`**: An array of weights for each data point. Each data point `x[i]` contributes `weights[i]` to the bin count.
*   **`color`**: Sets the color of the histogram bars.
*   **`edgecolor`**: Sets the color of the outline of the bars.
*   **`alpha`**: Controls the transparency of the bars (0.0 to 1.0).

**Example:**

```python
plt.hist(x = df["Age"],bins = 5, color="skyblue", edgecolor="black" ) # Its directly .hist() dont provide "kind" arg here
```

```python
# Assuming 'df' is your DataFrame with an 'Age' column

df['Age'].plot(kind='hist', bins=20, color = "pink", edgecolor='white') # White edge looks cool
```

1. B) Seaborn's `sns.histplot()`

**Parameters :**

  data : pandas.DataFrame, numpy.ndarray, mapping, or sequence

  x, y : cols to consider.

  bin : no of bins to group the data into.

  hue : Second Column to be considered as Inner category.

  kde : To show smooth curve using Kernel Density Estimate


```python

sns.histplot(data=df, x='Age', kde=True)
```

We can see in all hist plots that more no of people are from 20-40,less from 0 to 20, even lesser from 40 to 60, and very few old people (greater than 60)

# 2. DistPlot ( Now changed to DisPlot)

It is similar plot to histogram, but with kde enabled and shown on top of it.

It is found only on seaborn.

Distplot is Deprecated Hence use displot with, kde = True

We can pass data_frame["col_name"] directly if we want incase we dont want to use data, x, y arg format

```python
sns.displot(data= df, x="Age", kde = True)
```

```python
sns.displot(df["Age"], kde = True)
```

In Seaborn displot is new interface common for 3 plots histplot,kdeplot, ecdfplot()

```python
sns.displot(df["Age"], kind="kde")
```

```python
sns.displot(df["Age"], kind="ecdf")
```

If we set kde= True, the plot will be in terms of Frequency/ counts. But if we do kind="kde" or "ecdf", The plot will be in terms of probabalities since they are PDF curves.
