# 1. Bivariate Analysis: Numerical to Numerical

When both variables are numerical, we can use scatter plots, line plots, or heatmaps to visualize the relationship between them.

**Scatter Plot:** Shows the relationship between two numerical variables. Each point on the scatter plot represents a pair of values from the two variables. Sns allows upto 5 variable analysis in 2D.

**Line Plot:** Used when one of the numerical variables represents time or a sequential order. It shows how the other numerical variable changes over time or sequence.

**Heatmap:** A graphical representation of data where values are depicted by color. In bivariate analysis, a heatmap can show the correlation between two numerical variables.

```python
# Scatter plot for numerical to numerical analysis
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

In Scatterplots, each point is an entity.

We can further analyse these entities with respect to their categories.ie, We can add one more variable under which we can observe these x,y relationship, specifically we use "hue" for it.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
```

We can further add another category variable for analysis using "style" arg which are shown as X or O depending on category. Ex : Smoker

```python
sns.scatterplot(x= "total_bill", y="tip", hue="sex", style="smoker", data=tips)
```

We can further add another variable as category using "size" arg where the X and O shapes representing an entity, their size varies according to the vals of feature given in size arg. ex : size feature of tips dataset, signiying no of people ate in group.

```python
sns.scatterplot(x= "total_bill", y="tip", hue="sex", style="smoker", size="size", data=tips)
```

# 2. Bivariate Analysis: Numerical to Categorical

When one variable is numerical and the other is categorical, we can use box plots, violin plots, or bar plots to compare the distribution of the numerical variable across different categories.

**Bar Plot:** Used to show the mean or median of a numerical variable for each category. In sns, there appears a black vertical line on on bars, indicating confidence intervals.

**Box Plot:** Displays the distribution of a numerical variable for each category, showing the median, quartiles, and potential outliers.

**Dis Plot/ KDE Plot:** KDE density based, dist plot used to analyse and compare PDF distribution of continus feature within each category. But its better to use kdeplot

**Violin Plot:** Similar to a box plot, but it also shows the probability density of the data at different values, giving a more detailed view of the distribution.

### Bivariate Analysis: Categorical to Categorical

When both variables are categorical, we can use count plots, grouped bar plots, or heatmaps to visualize the relationship between the categories.

**Count Plot:** Shows the counts of observations in each category.

**Heatmap:** Can be used to visualize the contingency table of two categorical variables, showing the frequency of each combination of categories.

**ClusterMap:** Visualizes the relationships between categories by clustering rows and columns based on similarity. It's a good way to see patterns in the co-occurrence of categories.

## 1. CountPlot

```python
# Count plot for categorical to categorical analysis (using a subset of the titanic data)
sns.countplot(x='survived', hue='pclass', data=titanic)
```

We can see that among the people that did not survive, most were from class 3. And also among those that survived most were from class 1.

--------
## 2. Heatmap

Note that in order for us to analyse categorical to -categorical features, We need to first obtain crosstab between those features.

### Crosstab (Cross-Tabulation)

A crosstab (or cross-tabulation) is a table that shows the frequency distribution of two categorical variables. It displays the number of observations that fall into each combination of categories.

In simpler terms, it's like counting how many times each pair of categories appears together in your data. For example, if you have 'Sex' (Male/Female) and 'Survived' (Yes/No) as categorical variables in the Titanic dataset, a crosstab would show you:

*   How many males survived
*   How many males did not survive
*   How many females survived
*   How many females did not survive

You can calculate a crosstab in pandas using the `pd.crosstab()` function. It takes two Series (the categorical columns) as arguments and returns a DataFrame representing the crosstab.

For example, to get the crosstab of 'survived' and 'pclass' from the titanic dataset, you would use:

```python
pd.crosstab( titanic["pclass"], titanic["survived"])
```

Thus we obtained, how many from each class survived and did not survive.

We can use heatmap to represent this further

```python
sns.heatmap(pd.crosstab( titanic["pclass"], titanic["survived"]),color= sns.color_palette("coolwarm"))
```

# PairPlot

A pairplot is a powerful visualization tool that creates a grid of scatter plots for all pairs of numerical variables in a dataset, and histograms or kernel density estimates on the diagonal(since on diagnals its a single same variable hence univariate). It's useful for visualizing the relationships between multiple variables simultaneously and identifying potential correlations or patterns.

**Uses of PairPlot:**

*   **Exploring relationships:** Quickly visualize the relationships between all pairs of numerical variables in a dataset.
*   **Identifying correlations:** Spot positive or negative correlations between variables.
*   **Detecting patterns:** Identify clusters, outliers, or other interesting patterns in the data.
*   **Comparing distributions:** See the distribution of each individual variable on the diagonal plots.

**How to use PairPlot for analysis:**

*   **Scatter plots:** Examine the scatter plots to see how pairs of variables are related. Look for linear relationships, curves, or other patterns.
*   **Histograms/KDEs:** Look at the diagonal plots to understand the distribution of each individual variable.
*   **Hue:** Use the `hue` argument to color the points based on a categorical variable. This helps to see how the relationships between numerical variables differ across categories.

You can use `sns.pairplot()` to generate a pairplot. For example, to create a pairplot of the `iris` dataset, you would use:

```python
sns.pairplot(data=iris)
```

But most importantly we can provide "hue" to it, which means it can give us variation of scatterplots within each category of categorical feature we chose as hue.

```python
sns.pairplot(data=iris, hue="species")
```

# Line Plot (Numerical to Numerical, often with Time Series)

A line plot is used to visualize the relationship between two numerical variables, particularly when one of the variables represents a sequential order, such as time. It connects data points with lines, showing the trend or change in the numerical variable over the sequence.

**Uses of Line Plot:**

*   **Time Series Analysis:** Line plots are fundamental for visualizing time series data, showing how a variable changes over time (e.g., stock prices, temperature, passenger counts over years).
*   **Tracking Trends:** Easily identify trends, seasonality, and cycles in data.
*   **Comparing Trends:** Plot multiple lines on the same graph to compare the trends of different categories or groups over time.
*   **Sequential Data:** Useful for any data where the order of observations is meaningful, not just time.

**How to use Line Plot for analysis:**

*   **Observe the trend:** Look at the overall direction of the line to understand if the variable is increasing, decreasing, or remaining stable over time or sequence.
*   **Identify patterns:** Look for recurring patterns like seasonality (regular peaks and troughs) or cycles.
*   **Compare lines:** If plotting multiple lines, compare their trends and see how they differ.
*   **Identify anomalies:** Notice any sudden spikes or drops that deviate from the overall trend, as these might indicate unusual events.

You can use `sns.lineplot()` to generate a line plot. For example, to visualize the passenger counts over the years in the `flights` dataset, you would first need to combine the 'year' and 'month' columns into a datetime format, then you can plot the 'passengers' against this new datetime column.

```python
# Lets analyse the tips data, plotting the no of passengers for each year as time series
```

```python
flights.head()
```

```python
# But as we can see its in (year, month) format. Lets add passengers in each year, by summing up each month data

flights.groupby("year")["passengers"].sum()
```

```python
# But pandas makes the year(grouping var) as row label(index). We dont want it, Instead
# we can use reset_index() to reset to original 0,1,2,3 type of index

flights.groupby("year")["passengers"].sum().reset_index()
```

```python
# Now lets pass the year and passengers to lineplot

flight_passengers_per_year = flights.groupby("year")["passengers"].sum().reset_index()

sns.lineplot(x="year", y="passengers", data=flight_passengers_per_year)
```

```python
# We can plot the same in pandas .plot(kind="line") without reset_index()

flights.groupby("year")["passengers"].sum().plot(kind="line")
```

```python
# In the snsplot we can do something interesting

sns.lineplot(x="year", y="passengers", hue="month", data=flights)
```
